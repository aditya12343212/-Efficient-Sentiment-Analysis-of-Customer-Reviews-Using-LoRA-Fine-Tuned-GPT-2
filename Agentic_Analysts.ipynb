```python
import os
from google.colab import userdata

os.environ["GROQ_API_KEY"] = userdata.get('Aditya')

from crewai import Agent, Task, Crew, Process

print("âœ… Ready! Warnings are harmless â€” let's build the agents.")
```

```python

GROQ_MODEL = "groq/llama-3.3-70b-versatile"  # You can change to "groq/llama3-8b-8192" for even faster results

# Agents - all use Groq
researcher = Agent(
    role='Senior Market Researcher',
    goal='Research the topic deeply with accurate, up-to-date facts and trends',
    backstory="You are a world-class researcher with deep knowledge of current industries and technologies.",
    llm=GROQ_MODEL,
    verbose=True,
    allow_delegation=False
)

analyst = Agent(
    role='Expert Data Analyst',
    goal='Analyze research findings to extract key insights, opportunities, and risks',
    backstory="You are a sharp analyst who identifies patterns, predictions, and actionable conclusions.",
    llm=GROQ_MODEL,
    verbose=True,
    allow_delegation=False
)

writer = Agent(
    role='Professional Report Writer',
    goal='Write clear, structured, and engaging professional reports',
    backstory="You are an award-winning writer specializing in business and technology reports.",
    llm=GROQ_MODEL,
    verbose=True,
    allow_delegation=False
)

reviewer = Agent(
    role='Senior Editor & Quality Reviewer',
    goal='Polish the report for clarity, accuracy, flow, and professionalism',
    backstory="You are a meticulous editor who ensures the final output is flawless and impactful.",
    llm=GROQ_MODEL,
    verbose=True,
    allow_delegation=False
)

# Tasks
research_task = Task(
    description="Research {topic} thoroughly. Cover current status, key trends, major players, statistics, challenges, and future outlook.",
    expected_output="A detailed research summary with key facts and insights.",
    agent=researcher
)

analysis_task = Task(
    description="Analyze the research findings. Identify the most important insights, opportunities, risks, and future predictions.",
    expected_output="A clear, structured analysis with bullet points.",
    agent=analyst
)

writing_task = Task(
    description="""Write a full professional market research report in markdown format with these sections:
- Executive Summary
- Market Overview
- Key Trends & Insights
- Major Players
- Opportunities & Challenges
- Future Outlook & Recommendations
- Conclusion""",
    expected_output="A well-formatted, engaging, and professional report in markdown.",
    agent=writer
)

review_task = Task(
    description="Review and edit the entire report. Improve clarity, grammar, flow, structure, and overall professionalism.",
    expected_output="The final polished and high-quality report.",
    agent=reviewer
)

# Create the Crew
market_research_crew = Crew(
    agents=[researcher, analyst, writer, reviewer],
    tasks=[research_task, analysis_task, writing_task, review_task],
    process=Process.sequential,
    verbose=True,
    memory=False  # Keeps it fast and avoids any issues
)

print("ğŸš€ Your FREE Groq-Powered Multi-Agent Crew is READY!")
```

```python
# Choose your topic here!
topic = "Future of AI agents in 2026"  # Change to anything you like, e.g., "Best AI projects for resume 2026" or "AI career in India"

print(f"ğŸ”¥ Starting ultra-fast research on: {topic}\n")

# Run the crew
result = market_research_crew.kickoff(inputs={'topic': topic})

# Display the final report
print("\n" + "="*80)
print("YOUR FINAL PROFESSIONAL REPORT")
print("="*80)
print(result)
```
